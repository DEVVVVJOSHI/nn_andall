```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rill_no=[1,2,3,4,5,6,7,8,9]
marks=[1,1,1,1,1,1,12,3,4]
sample_df= pd.DataFrame({"rill_no":rill_no,"marks":marks})
sample_df.head()
```

LINE PLOT

```python
sns.lineplot(x='rill_no',y='marks',data=sample_df)
plt.title('student marks')
```

```python
seaboen_df=sns.load_dataset('titanic')
seaboen_df.head()
```

```python
sns.displot(x='age',y='class',data=seaboen_df) #displot plots bar graph
```

histplot


```python
sns.histplot(x='age',y='class',data=seaboen_df)
```

```python
bins=[1,2,3,4,5]
sns.histplot('fare',bins=bins,fill=False,kde =True)
plt.xticks(bins) #xtick ticks the x axis of the graph in matplotib
```

SCATTERPLOT

```python
sns.scatterplot(x='age',y='fare',hue='alive',style='class',palette='inferno',alpha=0.5,data=seaboen_df)#hue is used to differenciate using a parameter
#alpha is used to set the transparency
```

BAR PLOT

```python

sns.barplot(x='age',y='fare',data=seaboen_df,hue='sex')
```

```python
gandu=sns.load_dataset('flights')
gandu.head()
```

heatmap

```python
gandu=gandu.pivot(index="month",columns="year",values="passengers")#index,col,values dalna imp hai abse
gandu.head()
```

```python
plt.figure(figsize=(10,10))
dev=sns.heatmap(gandu,annot=True,fmt='d',cmap='PuBu',cbar_kws={"orientation":"horizontal"})#cbar_kws sets orient
dev
```
